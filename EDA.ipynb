# Exploratory Data Analysis

Loading necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
```

```python
data = pd.read_csv("list_attr_celeba.csv")     #Loading the Labels
```

```python
data.head()    #Jist of Attributes
```

```python
#Converting -1s to 0s
def func(x):
    if(x==-1):
        return 0
    else:
        return x
for k in data.columns:
    data[k] = data[k].apply(func)
```

```python
data.head(15)
```

### Sample Images and corresponding Attributes

```python
path = r'img_align_celeba//img_align_celeba//'
for i in range(9,12):
    print("Image:")
    image = cv2.imread(path+data.iloc[i]["image_id"])
    rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb_img)
    plt.show()
    print("Labels:")
    col_list = list(data.columns[1:])
    for a in col_list:
        if (data.iloc[i][a] == 1):
            print(a)
    print("-----------------------------------------")
    
```

```python
#Statistics of the data
stats = data.describe()
stats
```

### Visualization of frequency of each Attribute

```python
vis=stats.iloc[1]
vis.columns=['Attributes','Mean']
plt.figure(figsize=(15,5))
plt.bar(vis.index,vis,color="green")
plt.legend(["Mean"])
plt.xticks(rotation=90)
plt.axvspan(-1,40,ymin=0.3,ymax=0.7, color='red', alpha=0.2)
plt.show()
```

We can see that only few features are Balanced like
1. Attractive
2. Heavy Makeup
3. High Cheekbones
4. Male
5. Mouth Slightly open
6. Oval Face
7. Pointy Nose
8. Smiling
9. Wavy Hair
10. Wearing Lipstick

Apart from these all other Attributes are considered as biased due to their low frequency


```python
#List of Features with low balance of positive class and negative class

mean_list = []
count=0
print("Features with low balance in positive class and negative class")
print()
for i in range(len(stats.iloc[1])):
    mean_list.append((stats.iloc[1].index[i],stats.iloc[1][i]))
    if(stats.iloc[1][i]<0.3 or stats.iloc[1][i]>0.7):
        count+=1
        print(stats.iloc[1].index[i],": mean -",stats.iloc[1][i])
        print("count of 1:",len(data[data[stats.iloc[1].index[i]]==1][stats.iloc[1].index[i]]))
        print("count of 0:",len(data[data[stats.iloc[1].index[i]]==0][stats.iloc[1].index[i]]))
        print()
        
print("Total:",count,"features")
```

```python
#List of Features with good balance of positive class and negative class
good_count=0
print("Features with good balance in positive class and negative class")
print()
for c,i in mean_list:
    if(i>=0.3 and i<=0.7):
        good_count+=1
        print(c,": mean -",i)
        print("count of 1:",len(data[data[c]==1][c]))
        print("count of 0:",len(data[data[c]==0][c]))
        print()
        
print("Total:",good_count,"features")
```

### Correlation of Attributes

```python
plt.figure(figsize=(12,12))
sns.heatmap(data.iloc[:,1:].corr(), cmap="viridis")
plt.show()
```

### Bounding Box Data - Analysis

```python
bbox_data = pd.read_csv("list_bbox_celeba.csv")
```

```python
bbox_data.head(7)
```

```python
path = r'img_align_celeba//img_align_celeba//00000'
for i in range(6,9):
    image = cv2.imread(path+str(i)+'.jpg')
    x = bbox_data.iloc[i-1][1]
    y = bbox_data.iloc[i-1][2]
    w = bbox_data.iloc[i-1][3]
    h = bbox_data.iloc[i-1][4]
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
    rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb_img)
    plt.show()
```

As the Bounding Box Data given in the Dataset are not the Bounding Boxes of the given Images, they are retrieved from original dataset. The Images given in the dataset are Cropped Images. So, the Bounding Boxes do not map to these Images.

### Facial Landmarks Data
